## Overview

The increase in supply and demand for electric vehicles (EVs) in the United States is significant change in the technology in the last decade. However, the availability of supporting charging infrastructure is not able to keep up in many rural areas across the country. Despite the increase in EV sales, the EV lack of commonly available EV charging infrastructure may be a reason for why road trips driven in EVs are not commonplace. However, EV charging infrastructure remains prevalent in urban and suburban areas. This uneven distribution of infrastructure may reinforce existing disparities between urban, suburban, and rural communities, making it ain interesting situation to spatially analyze.

The primary goal of this project is to analyze the spatial distribution of EV charging stations across the United States, with a focus on differences between urban, suburban, and rural areas. This project aims to identify various EVs' spatial patterns such as including clustering and dispersion. More broadly, the project seeks to understand how infrastructure availability varies geographically and to provide insight into patterns of accessibility across different types of regions. These goals should aim to fully connect various urban domains together, rather than simply analyzing just one or two datasets.

The main hypothesis is that EV charging stations show positive spatial autocorrelation, meaning they are concentrated in densely populated and higher-income metropolitan areas, while rural and lower-income regions have a much lower count of charging stations. A secondary hypothesis is that EV charging station density is positively associated with socioeconomic variables such as median household income. A common incentive to purchace an electric vehicle is that one can avoid paying for gas, despite its initial average higher cost than gas vehicles. Additionally, another hypothesis is that environmental variables such as air quality, could be positively associated with EV charging station locations. Potential variations or bias in the data could include the data generating process (DGP) being difficult for collecting data for rural charging stations. Any publicly available datasets may only include mainline charging stations such as Tesla. There may be many other manufacturers of charging stations that are not recorded. The DGP might also vary by state. Some states may make some data private.

## Proposed Methods

This project will integrate multiple datasets to conduct a spatial analysis of EV charging infrastructure in the United States. It will contain many statistical tests such as analyzing spatial autocorrelations and exploratory visualizations.

### Data

This project will use four main datasets. The primary dataset is from the NREL, which has point data of alternative fuel charging stations. It is being accessed via the website of US Department of Energy. The key attributes of interest are station location, network provider, charger level (Level 1, Level 2, or DC Fast Charge), and number of available ports.

The second dataset is from the U.S. Census Bureau's American Community Survey (ACS) 5-Year Estimates, accessed through the Census Bureau's data portal, which is also where the data is also originally from. The key attributes of interest are median household income and educational attainment, both of which serve as socioeconomic indicators for multivariate spatial analysis.

The third dataset is the U.S. Census Bureau TIGER/Line Shapefiles, which is also originally from the US Census Bereau. The key attribute of interest is the county boundary geometry, which serves as the spatial unit of analysis to which all other datasets will be joined.

The fourth dataset is the U.S. Environmental Protection Agency's Air Quality Index (AQI) Annual Summary, accessed through the EPA's AirData portal, where the data is also originally from. structured as a tabular CSV file. The key attribute of interest is the annual AQI value at the county level, which serves as an environmental indicator for spatial analysis.

### Statistical Summary

The statistical summaries will be generated via Python, using the `pandas` and `geopandas` libraries. All summaries will be computed at the county level. County aggregation may be the most appropriate for this project, as this analysis is on a national scale. Although an aggregation such as census tracts would be more accurate than counties, it may not be necessary for a national scale analysis. For the NREL dataset, stations will be filtered to EV charging stations only and aggregated to the county level by count. The distribution of this variable will be characterized by its mean, median, standard deviation, minimum, maximum, and right versus left skewdness. For the ACS dataset, median household income and educational attainment will be summarized using mean, median, standard deviation, minimum, maximum, and interquartile range (IQR). For the EPA Air Quality dataset, the county-level AQI values will be summarized using the same set of statistics. All variables will be visualized using histograms and boxplots via the `matplotlib` and `seaborn` libraries to assess distributional shape and identify outliers.

### Spatial Analysis

The first step will be statistical mapping via choropleth maps, produced using `geopandas`, `seaborn` and `matplotlib`. Choropleth maps will be generated for each key attribute: EV station count per county, median household income by county, mean education level by county, and mean AQI bu county. These maps are meant to establish a visual baseline for identifying patterns and correlations before any more sopphisticated analysis methods are performed.

The second step will be a test for spatial autocorrelation, so a global Moran's I test will be calculated for EV station counts to determine whether the spatial distribution of charging infrastructure is clustered, dispersed, or random across U.S. counties. A spatial weights matrix will be used to define spatial neighbors between counties. If the global Moran's I outputs any statistically significant clustering, local Moran's I (LISA) will then be calculated for further precision than the global Moran I's test. Visualizations for theese tests will then be generated.

The third step will be exploratory spatial data analysis. Another Moran's I test will be calculated to assess the spatial correlation between EV station counts and each of the supplementary variables: median household income, mean educational level, and mean AQI. This will be used to test the hypothesis that counties with high EV infrastructure tend to be clustered near counties of higher levels of these three supplementary variables. These methods directly address the research hypotheses and are appropriate given the county-level, areal data structure of the analysis.

This proposed analysis will be done because EV charging infrastructure is unlikely to be randomly distributed, and spatial autocorrelation testing allows for results more sophisticated than simple correlation analysis.
